```python
# Criando função para separar o arquivo em um arquivo por sequência

def separa_fasta(arquivo):
  with open(arquivo, 'r') as arquivo:  # abrindo o arquivo em modo de leitura
    sequencia = ''
    nome_sequencia = ''
    contador = 1


    for linha in arquivo:
      linha = linha.strip() # exclui espaços, quebras de linha e tabulações
      if linha.startswith('>'):
        # gravar a sequência anterior se existir
        if sequencia:
          # limpar caracteres inválidos para o nome do arquivo
          nome_arquivo = nome_sequencia.replace('.','_')
          with open(f'{nome_arquivo}.fasta', 'w') as output_file:
            output_file.write(f'>{nome_sequencia}\n{sequencia}\n')
          sequencia = ''
          contador += 1     # contador = contador + 1

        # atuliza o nome da nova sequência
        nome_sequencia = linha[1:] # remove o '>'
      else:
        sequencia += linha

      # Grava a última sequência
      if sequencia:
        nome_arquivo = nome_sequencia.replace('.','_')
        with open(f'{nome_arquivo}.fasta', 'w') as output_file:
          output_file.write(f'>{nome_sequencia}\n{sequencia}\n')
```
